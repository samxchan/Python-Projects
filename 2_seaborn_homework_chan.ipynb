```python
# Import the Palmer penguins dataset into a pandas DataFrame
df = sns.load_dataset('penguins')

# Convert to category data type
df[['species', 'island', 'sex']] = df[['species', 'island', 'sex']].astype('category')

# Drop rows with missing values
df = df.dropna().reset_index(drop=True)

# Look at the first 5 rows of the data
df.head()
```

# Line Plot

Here, you should use the seaborn library to create a seaborn line chart, with the `"bill_length_mm"` column along the x-axis and the `"bill_depth_mm"` column along the y-axis. Remember to add a title to your chart and label the x- and y-axis!

# Count Plot

Seaborn's `countplot()` method can be thought of as a histogram across a categorical field, instead of a quantitative variable. 

Here, you should use the `countplot()` method to find the answer to the following question.

**Question:** Which island has the most penguins?  

**Answer**: Biscoe

`seaborn.countplot` [documentation](https://seaborn.pydata.org/generated/seaborn.countplot.html#seaborn.countplot)

```python
# Create a count plot using the "island" field
ax = sns.countplot(x='island', data=df)
ax.set_title("Number of Penguins by Island")
ax.set_ylabel("Number of penguins")
```

To answer the next question, this time you should take advantage of the `countplot()` function's optional `hue` parameter. In addition, you will also need to:
- Plot the bars horizontally
- Add a title and label the x- and y-axis
- Select one seaborn's optional color palettes: `"deep"`, `"muted"`, `"pastel"`, `"bright"`, `"dark"`, `"colorblind"`

**Question:** On Biscoe island, which sex is predominate?  

**Answer:** Male

```python
# Create a countplot
ax = sns.countplot(data=df, hue = 'sex', y = 'island', palette = 'deep')
ax.set_title("Penguin gender by Island")
```

# Cat Plot

Seaborn's `catplot()` is a figure-level interface that provides access to several axes-level functions that we can use to show the relationship between a numerical variable and one or more categorical variables in our dataset.

To answer the following question, you should use `catplot` to create a count plot. If you forget how to do this, be sure to check out the documentation.

**Question:** Do any of the islands only have one species of penguin?   

**Answer:** Torgersen

`seaborn.catplot` [documentation](https://seaborn.pydata.org/generated/seaborn.catplot.html)

```python
# Create a countplot using catplot()
g = sns.catplot(data=df, x='island', height=5, ci=False, kind='count', hue='species')
#g.set_titles("Number of Penguins by Island")
g.set_titles("Number of Penguins by Island")
```

# Histogram

Similar to `catplot`, seaborn's `displot()` method is a figure-level interface to several axes-level functions such as `histplot()` and `kdeplot()`. 

Here, you should use `displot()` to visualize the distribution of the `"flipper_length_mm"` variable. Hint: remember that you can see more detail by increasing the number of bins.

Next, you should briefly describe in no more than one or two sentences whether or not you think the feature is normally distributed. 

**Answer:** I believe the feature is not normally distributed. It has some features of a normal distribution; it is bell-shaped and has one peak. However, it is not symmetric around the mean.

`seaborn.displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
# Use displot() to plot a histogram
g = sns.displot(df, x='flipper_length_mm')
g.set_titles("Number of Penguins by Flipper Length")
g.set_xlabels("Flipper Length")
```

Here, you are interested in learning more about the distribution of the penguins' flipper lengths. Using `displot()` again, this time pass in the `"species"` variable to the `hue` parameter so it will draw a separate histogram for each of its unique values and distinguish them by color.

```python
# Use displot() to plot a histogram
g = sns.displot(df, x='flipper_length_mm', hue = 'species')
g.set_titles("Number of Penguins by Flipper Length")
g.set_xlabels("Flipper Length")
```

That last plot was a little difficult to read! However, since `displot()` is a figure-level function, we can use it to draw each individual distribution in a separate subplot for each unique species. If you don't remember which optional parameter to use, just check out the `displot()` documentation!

`seaborn.displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
# Use displot() to plot a histogram
g = sns.displot(data = df, x='flipper_length_mm', hue = 'species', col="species")
g.set_titles("{col_name} penguins")
```

# Scatter Plots

Here, you should start by using seaborn to create a scatter plot to see if you can detect a relationship between the `"bill_length_mm"` and `"bill_depth_mm"` columns. Does there appear to be a positive or negative correlation between the columns?

Next, look to see if you can detect a relationship between a penguin's flipper length and their body mass. To do this you should create a scatter plot again; however, this time you should use the figure-level `relplot()` function.

# Regression

Using `lmplot()`, here you should try to find if there is a relationship between the penguins' body mass and bill lengths when looked at by species.

**Question:** Is there a relationship between the penguins' body mass and bill length when looked at by species?  

**Answer:** There is a positive relationship between body mass and bill length for each species

  
`seaborn.lmplot` [documentation](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

```python
# Use lmplot() to see if there is a relationship
g = sns.lmplot(data=df, x='body_mass_g', y='bill_length_mm', hue='species')
g.set_titles("Body mass vs Bill length by species")
```

# Box Plots

Now, here you will need to look at how the penguins' flipper length varies by species by using seaborn's axes-level `boxplot()` function to answer the question.

**Question:** Which species has the longest flippers and which species has the shortest?

**Answer:** Gentoo penguins have the longest flippers and Adelie penguins have the shortest.

`seaborn.boxplot` [documentation](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

```python
# Create a box plot
ax = sns.boxplot(data=df, y='flipper_length_mm', x = 'species')
ax.set_title("Flipper length by species")
```

Now, you want to create a similar box plot, but this time looking at how the penguins' body mass varies by species. 

**Question:** Which species is the heaviest?  

**Answer:**  Gentoo penguins are the heaviest.

```python
# Create a horizontal box plot
ax = sns.boxplot(data=df, x='body_mass_g', y='species', orient = 'h')
ax.set_title("Body mass by species")
```
